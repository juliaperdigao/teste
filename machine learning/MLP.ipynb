```python
# imported pandas library to use DataFrames
import pandas as pd

# open the csv file as a Dataframe
df = pd.read_csv("student-mat.csv", sep = ";")

# rename the columns to lower case sentence
df.columns= df.columns.str.lower()

# visualize DataFrame "head" to confirm it was correctly generated
df.head()
```

```python
# make a copy of our data set so that we can prepare it for our k-NN classification
data = df
```

```python
# put outcome in its own object
g3_outcome = df[['g3']].copy()

# remove original variable from the data set
data.drop('g3', axis='columns', inplace=True)

# visualize DataFrame "head" to confirm it was correctly generated
data.head()
```

```python
# determine which variables are integers
data.info()
```

```python
# The variables:
# “age” 
# “medu” 
# “fedu” 
# “traveltime” 
# “studytime” 
# “failures” 
# “famrel” 
# “freetime” 
# “goout” 
# “dalc” 
# “walc” 
# “health” 
# “absences” 
# “g1” 
# “g2” 
# are interger variables, which means they can be scaled

# import StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# scale the variables
data [["age", "medu", "fedu", "traveltime", "studytime", "failures", "famrel", "freetime", "goout", "dalc", "walc", "health", "absences", "g1", "g2"]]= scaler.fit_transform(data[["age", "medu", "fedu", "traveltime", "studytime", "failures", "famrel", "freetime", "goout", "dalc", "walc", "health", "absences", "g1", "g2"]])

# visualize DataFrame "head" to confirm it was correctly generated
data.head()
```

```python
# import LabelEncoder
from sklearn import preprocessing


# creating labelEncoder
le = preprocessing.LabelEncoder()


# converting string labels into numbers
data['schoolsup']=le.fit_transform(data['schoolsup'])
data['famsup']=le.fit_transform(data['famsup'])
data['paid']=le.fit_transform(data['paid'])
data['activities']=le.fit_transform(data['activities'])
data['nursery']=le.fit_transform(data['nursery'])
data['higher']=le.fit_transform(data['higher'])
data['internet']=le.fit_transform(data['internet'])
data['romantic']=le.fit_transform(data['romantic'])

 
# then dummy code variables that have two levels, but are not numeric
data['school']=le.fit_transform(data['school'])
data['sex']=le.fit_transform(data['sex'])
data['address']=le.fit_transform(data['address'])
data['famsize']=le.fit_transform(data['famsize'])
data['pstatus']=le.fit_transform(data['pstatus'])
```

```python
# dummy code variables that have three or more levels:
# "mjob"
# "fjob"
# "guardian"
# "reason"

# dummy "mjob"
dummy_mom = pd.get_dummies(data['mjob'])

# rename the columns so we can incorporate them in the DataFrame (data) later
dummy_mom = dummy_mom.rename(columns = {"at_home": "MOM_at_home", 
                                  "health":"MOM_health", 
                                  "other": "MOM_other",
                                  "services":"MOM_services", 
                                  "teacher":"MOM_teacher" }) 

# visualize DataFrame "head" to confirm it was correctly generated
dummy_mom.head()
```

```python
# dummy "fjob"
dummy_dad = pd.get_dummies(data['fjob'])

# rename the columns so we can incorporate them in the DataFrame (data) later
dummy_dad = dummy_dad.rename(columns = {"at_home": "DAD_at_home", 
                                  "health":"DAD_health", 
                                  "other": "DAD_other",
                                  "services":"DAD_services", 
                                  "teacher":"DAD_teacher" }) 

# visualize DataFrame "head" to confirm it was correctly generated
dummy_dad.head()
```

```python
# dummy "guardian"
dummy_guardian = pd.get_dummies(data['guardian'])

# rename the columns so we can incorporate them in the DataFrame (data) later
dummy_guardian = dummy_guardian.rename(columns = {"other": "other_guardian"}) 

# visualize DataFrame "head" to confirm it was correctly generated
dummy_guardian.head()
```

```python
# dummy "reason"
dummy_reason = pd.get_dummies(data['reason'])

# rename the columns so we can incorporate them in the DataFrame (data) later
dummy_reason = dummy_reason.rename(columns = {"other": "other_reason"}) 

# visualize DataFrame "head" to confirm it was correctly generated
dummy_reason.head()
```

```python
# remove original variables that had to be dummy coded
data.drop(["mjob", "fjob", "guardian", "reason"], axis=1)
```

```python
# combine new dummy variables with original data set
data_ready = pd.concat([data.drop(["mjob", "fjob", "guardian", "reason"], axis=1), dummy_mom, dummy_dad, dummy_reason, dummy_guardian], axis = 1)


# visualize DataFrame "head" to confirm it was correctly generated
data_ready.head()
```

```python
# transform the outcome column in a binary 'yes or no' column
# this step is recquired so we can decide if the student was approved or not in Math class
# 'yes' means he was approved
# 'no' means the opposite
# so the classes are "APPROVED" represented by 'yes' and "NOT APPROVED" represented by 'no'

# the rule is: if the value in former 'g3' column was less than or equal to 10, the student is not approved
for i in range(len(g3_outcome)):
    if g3_outcome.loc[i, 'g3'] > 10:
        g3_outcome.loc[i, 'g3'] = 'yes'
    else:
        g3_outcome.loc[i, 'g3'] = 'no'
```

```python
# split the data into training and test sets 
# we partition 75% of the data into the training set and the remaining 25% into the test set

# create an 'X' matrix by dropping the irrelevant columns
X = data_ready
y = g3_outcome

# import train_test_split function
from sklearn.model_selection import train_test_split

# split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(activation='tanh', alpha=0.05, max_iter=500, solver='sgd')
mlp.fit(X_train, y_train)
mlp.score(X_test, y_test)
```
